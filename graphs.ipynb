```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [1, 2, 4]
labels = ['Apple', 'Banana', 'Orange']
width = 0.4

fig, ax = plt.subplots()
labels = ['label 1', 'label 2', 'label 3']
ax.bar(x, y1, width=width, label="y1")

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label="y2")
ax.legend()

label_axis = [num + width/2 for num in x]
plt.xticks(label_axis, labels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fix, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
y3 = [2, 6, 8]
y1_y2_total = np.add(y1, y2)
y_total = np.add(y1_y2_total, y3)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x, y_total, tick_label=labels, label='y3')
ax.bar(x, y1_y2_total, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.DataFrame([['', 5],
                   ['', 15],
                   ['', 25],
                   ['SNS', 50],
                   ['', 1],
                   ['', 4]], columns = ["", "%"])
df

labels = df.loc[:, ""]
rate = df.loc[:, "%"]

fig, ax = plt.subplots()
explode = [0.2, 0, 0, 0, 0, 0]
ax.pie(rate, labels=labels, autopct='%1.0f%%', explode=explode)
ax.axis('equal')
plt.show()
```
